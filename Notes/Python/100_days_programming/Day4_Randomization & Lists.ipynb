```python
Common list methods:

                                                               Adding elements~
1. Append(value) -- adds element at the end.
         Ex: List = [1,2,3]
            List.append(4)     #O/p --> [1,2,3,4]

2. Insert(index, value) -- adds element at mentioned index.
         Ex: List = [1,2,3]
             List.Insert(2,4) #O/P -> [1,2,4,3]

3. Extend(iterable) -- extends the existing list.
         Ex: List = [1,2,3]
             List.extend([4,5,6])  #O/P --> [1,2,3,4,5,6]

                                                              Removing elements~
4. Pop(index) -- removes and return the eleement.
         Ex: List = [1,2,3,4]
             List.pop()    #removes last index value. --> [1,2,3]
             List.pop(2)   #removes index 2 value. --> [1,2,4]

5. remove(value) -- remove the first occurance value from a string.
         Ex: list = [1,2,3]
             list.remove(3)  #O/p --> [1,2]

6. Clear() -- removes all elements from the list.
         Ex: list = [1,2,3]
             list.clear()  #O/P --> []

                                                              Searching elements~
7. index(value) -- return the index of first occurance of the list.
         Ex: list = [1,2,3,2]
             list.index(2)  #O/p --> 3

8. count(value) -- count occurance of list.
         Ex: list = [1,2,3,2,5,6,2,7,9,2]
             list.count(2)     #O/p --> 4

9. In keyword -- check is that present in the list.
         Ex: list = [1,2,3,4]
             print(3 in list)    #True.

                                                              Sorting Elements~
10. sort() -- sorts the list in place.
         Ex: list = [9,7,5,2]
             list.sort()      #O/p --> [2,5,7,9]

11. sorted(iterable) -- returns a new sorted list.
         Ex: list = [9,5,1,8,4]
             list1 = sorted(list)   #O/p -->[1,4,5,8,9]

12. reversed(iterable) -- returns an iterator for the reversed list.
         Ex: list = [3,5,6,8,9,4,2,7]
             list.reverse()          #O/p --> [7, 2, 4, 9, 8, 6, 5, 3]

13. copy() -- create a shallow copy
         Ex: list = [1,2,3,5,6]
             list1 = list.copy()     #O/p --> list1 = [1,2,3,5,6]

14. [:] -- slicing the list and also creates a copy.
```

````python
#Day 4 Project: Rock Paper Scissors Game
import random
print("Welcome to the Rock Paper Scissors Game!")

# Rock
rock = """
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
"""

# Paper
paper = """
     _______
---'    ____)____
           ______)
          _______)
         _______)
---.__________)
"""

# Scissors
scissors = """
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
"""

options = [rock, paper, scissors]

#user input
user_choice = int(input("What do you choose? Type 0 for Rock, 1 for Paper or 2 for Scissors.\n"))
print("User Choice:")

#Checking user response valid or invalid.
if user_choice >= 0 and user_choice <= 2:
    print(options[user_choice])

    #computer generated choice
    computer_choice = random.randint(0,2)
    print("Computer Choice:")
    print(options[computer_choice])


#game logic
if user_choice >= 3 or user_choice < 0:
    print("Invalid choice! You lose.")
elif user_choice == computer_choice:
    print("It's a draw!")
elif user_choice == 0 and computer_choice == 2:
    print("You win!")
elif user_choice == 2 and computer_choice == 0:
    print("You lose!")
elif user_choice > computer_choice:
    print("You win!")
elif user_choice < computer_choice:
    print("You lose!")
````
