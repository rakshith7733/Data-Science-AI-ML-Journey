````python
#STEP 1: Picking a Random words and checking.

import random
word_list = ["cake","rakshith","popcorn","mobile","concert","books"]

#1 Randomly select a word from the list
random_word = word_list[random.randint(0,len(word_list)-1)]
print(random_word)

#2 Ask the user to guess the word and make the guess in lowercase.
guess = input("guess the word: ").lower()
print(guess)

#3 Check if the guess is correct and print "correct" or "incorrect" for each letter in the guess.
lst = []
for word in random_word:
    if guess == word:
        lst.append(word)
        print("correct")
    else:
        print("incorrect")
print(lst)
````

````python
#STEP 3: Checking if the player has won or lost.

import random
word_list = ["cake","rakshith","popcorn","mobile","concert","books"]

random_word = random.choice(word_list)
print(random_word)                      

game_over = False
correct_guess = []
            

placeholder = ""

for position in random_word:
    placeholder += "_"
print(placeholder)


while not game_over:
    guess = input("guess the word: ").lower()

    display = ""
    for letts in random_word:
        if letts == guess:
            display += letts
            correct_guess.append(letts)
        elif letts in correct_guess:
            display += letts
        else:
            display += "_"
    print(display)

    if display == random_word:
        game_over = True
        print("Game Over! You win!")
        
````
