Type error:
           An exception that occurs when the data types of an object in an operation is inappropriate. this occurs when the operations performed on the object is incorrect or not supported for the object.

Value error:
           when an object is given the incorrect value, the interpretor will raise ValueError. This occurs when the value is incorrect for the intended operations or if the value is missing altogether.

```python
#BMI Calculator
height = 1.65
weight = 84

bmi = round(weight / (height * height),2)
print(f'Your BMI Score is {bmi}')
```

```python
#Day2 Project: Tip Calculator
print("Welcome to the tip calculator.")

total_bill = float(input("What was the total bill? $"))
tip_percentage = int(input("What percentage tip would you like to give? 10, 12, or 15? "))
number_of_people = int(input("How many people to split the bill? "))

tip_amount = total_bill * (tip_percentage / 100)
total_amount = total_bill + tip_amount
amount_per_person = round(total_amount / number_of_people,2)
print(f"Each person should pay: ${amount_per_person:.2f}")
```
