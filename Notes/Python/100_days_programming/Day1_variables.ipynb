Syntax: The set of rules that defines how a python program will be written and interpreted.

Error Types:

Syntax Error: This occurs when the interpretor encounters invalid syntax in the code.
    EX:. Missing, mispelled, keyword, incorrect indentation, missing parantheses and so on..

NameError: A variable or function that hasn’t been defined.
    Ex:. num = 5
         print(x)   #printing variable which is not defined.

Indentation Error: Python expects proper spacing (indentation), especially in blocks like if, for, def.
    Ex:. if True:
         print(x)  #missing indentation.

Runtime Error: Code runs but crashes during execution.
    Ex:. x = 10 /0
         Print(x)

Type Error: These errors occur when you are using the wrong data types, invalid type annotations.
    EX:. len(1234)

Index Error: You are accessing an index that doesn’t exist.
    Ex:. arr = [1, 2, 3]
         print(arr[5])    #asking out of range.
