```python
Function:
        A function is a wrapper for a block of code. when you call it by name, all the codes/tasks within the function block will be executed.
        It helps us save time and repeated code.

# Defining a new function:

Def <function_name>():
      #Do this
      #then do this
      #finally do this
      #do something else.....

# Calling a function: 
                    This will trigger the function to perform the tasks as under function. We can call a function at any point of our code.

<function_name>()
```

While Loop:
           Until the certain condition is true perform the action.

Syntax:

While something_is_true:
      #Do something repeatedly
