```python
Nested IF-Else statement:

          do this
       else:
          do this
    else:

```

````python
#Rollercoster ride
#Condition height >= 120 & age less than 12, 12-18 and over 18.
print("Welcome to the rollercoaster!")
height = int(input("Enter your height? in CM?"))

if height >= 120:
    print("You can ride.")

    age = int(input("Enter your age?"))
    if age < 12:
        print("You need to pay $5.")
    elif age <= 18:
        print("You need to pay $7.")
    elif age > 18:
        print("You need to pay $12.")
        
else:
    print("Can't ride.")
````

````python
#Rollercoster ride
#Condition height >= 120 & age less than 12, 12-18 and over 18. add need picture of ride then add $3. Aged ticket(age = 45-55) will cost 0 for ride.
print("Welcome to the rollercoaster!")
height = int(input("Enter your height? in CM?"))

bill = 0

if height >= 120:
    print("You can ride.")

    age = int(input("Enter your age?"))
    if age < 12:
        bill = 5
        print("You need to pay $5.")
    elif age <= 18:
        bill = 7
        print("You need to pay $7.")
    elif age > 18 and age < 45 or age > 55:
        bill = 12
        print("You need to pay $12.")
    else:
        bill = 0
        print("Your ticket is free.")
    
    photo = input("Do you want a photo taken? Y or N.").lower()
    if photo == "y":
        bill += 3
    
    print(f"Your final bill is ${bill}.")
````

```python
~ Pizza Delivery.

Pizza scenario requirements:
1. small pizza = 15, medium pizza = 20, large pizza = 25
2. pepperoni for small = 2, pepperoni for medium & large = 3
3. extra cheese = 1 for any pizza
```

````python
print("Welcome to Python Pizza Deliveries!")
size = input("What size pizza do you want? S, M, or L: ").lower()
pepperoni = input("Do you want pepperoni on your pizza? Y or N: ").lower()
extra_cheese = input("Do you want extra cheese? Y or N: ").lower()

bill = 0

if size == "s":
    bill += 15
    if pepperoni == "y":
        bill += 2

elif size == "m":
    bill += 20
    if pepperoni == "y":
        bill += 3
elif size == "l":
    bill += 25
    if pepperoni == "y":
        bill += 3
else:
    print("Invalid size.")

#Cheese Option:
if extra_cheese == "y":
    bill += 1

print(f"Your final bill is: ${bill}")
````
